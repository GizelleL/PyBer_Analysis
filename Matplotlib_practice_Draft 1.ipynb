```python
#In this cell we are importing the matplotlib dependency
#we are chaining the pyplot module using matplotlib.pyplot which is the same as typing from matplotlib import pyplot
# Import dependencies.
import matplotlib.pyplot as plt
#the pyplot module helps us create all the neccessary charts; create a plotting area in a figure; plot lines in a plotting area and annotate the chart with labels, colors and decorative features
```

```python
#the below data is in the form of arrays(lists) 
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#in this cell we are chaining the plot() function to plt - what does chaining mean?
#the purpose of plt.plot() is to create the figure with the current axes
## Create the plot
plt.plot(x_axis, y_axis)
```

```python
#when we switch from using MATLAB to object oriented interface method we use ax.plot() inSTEAD OF PLT.PLOT()
fig, ax = plt.subplots()
ax.plot()
ax.plot(x_axis,y_axis)
```

```python
#To create a graph using the object oriented interface method we use the below
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#to change figure-level attributes (such as axis labels, a title or a legend) or save the figure as an image we use fig = plt.figure()
```

```python
#The plt.show() function looks for all the active figure objects and opens a window that displays them if you have two or more sets of data to plot
#this is how we use plt.show() function
#create the plot
plt.plot(x_axis, y_axis)
plt.show()
#plt.show() should only be used once per session because it can cause the graph to diplay improperly. If you have issues you may need to restart the kernel
```

```python
#When we annotate line charts, some of the features to make it more informative and visually appealing includes 
#distinct color and legends for each dataset that the lines or bars represent. 
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
#The above code creates the chart below: 
```

```python
#We can change the width and color of the line and add markers for data points. This is done by declaring parameters in 
#the plt.plot() function
#To declare the parameters for line color, width and marker we use color=, width = and marker= inside the plt.plot() function
#See sample code below: 
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis,y_axis,'o')
```

```python
#Make the markers larger by iterating through the y-axis data and multiply each data point by 3 and add it to a new list
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
#We can use the new y-axis list for the s parameter
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
#Format for list comprehension
#new_list = [expression for item in list if conditional]
#Below we are multiplying each data point in the y-axis list
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Making pie charts with the simple approach
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#We can add an "explode" to pie charts
#0 is the distance form the center of the pie, and "1" is completely outside the diameter of the pie
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
#Change the colors of the pie chart
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#Skill Drill Part 1- create a pie chart using the complex format
#Add a percentage to one decimal place to each wedge of the pie
#Increase the figure size to 8x8
fig,ax = plt.subplots()
ax.pie(y_axis,labels=x_axis,autopct='%.1f%%',radius=8,textprops={'fontsize': 48})
plt.show()
```

```python
#Skill Drill Part 2
#Explode the 2 highest percentage months
#Add a shadow
#How do I make the shadow more visible? - ask TA
fig,ax = plt.subplots()
explode_values = (0,0,1,0,0,0,1,0,0,0,0,0)
ax.pie(y_axis,labels=x_axis,autopct='%.1f%%',radius=8,textprops={'fontsize': 32},explode=explode_values,shadow=True,startangle=90)
plt.show()
```
